```python
import pandas as pd
import numpy as np
import json as js
from nltk import ngrams
from collections import defaultdict
```

```python
# buka file json
with open('./data/data.json','r') as file:
    json_data = js.load(file)

data = json_data['data']
```

```python
kata = []
for item in data:
    kata.append(item['kata'])
kata[:4]
```

```python
# membuat kamus bigram
bigramModel = defaultdict(lambda: defaultdict(lambda:0))
# melakukan perhitungan bigram probabilitas
for sentence in kata:
    sentence = sentence.split()
    for w1, w2 in ngrams(sentence, 2):
        bigramModel[w1][w2] += 1
```

```python
bigramModel
```

```python
# normalisasi probabilitas
for w1 in bigramModel:
    totalCount = float(sum(bigramModel[w1].values()))
    for w2 in bigramModel[w1]:
        bigramModel[w1][w2] /= totalCount
```

```python
bigramModel
```

```python
numStates = 2
transisiMatrix = np.array([[0.7,0.3],[0.4,0.6]])
initialProbabilitas = np.array([0.6,0.4])
```

```python
#penggunaan metode HMMs
def prdiksi_next_wort(prefWord):
    nextWordProbability = bigramModel[prefWord]
    state=np.argmax(initialProbabilitas)
    nextWord = max(nextWordProbability,key=lambda w: nextWordProbability[w]*transisiMatrix[state][state])
    return nextWord
```

```python
previousWord = "jangan"
next_word = prdiksi_next_wort(previousWord)
print("kata brikutnya dari {} adalah : {}".format(previousWord,next_word))
```
